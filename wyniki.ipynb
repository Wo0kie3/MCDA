```python
import pandas as pd
import os
import json

# Ścieżki do głównych katalogów
input_dir = 'rankings_and_matrix'  # Aktualizuj do swojej ścieżki
results_dir = 'results_fixed'      # Aktualizuj do swojej ścieżki

# Słowniki do przechowywania wyników
electre_binary_matrix_data = []
promethee_matrix_data = []
electre_matrix_data = []

electre_binary_complete_results = []
electre_binary_partial_results = []
electre_complete_results = []
electre_partial_results = []
promethee_complete_results = []
promethee_partial_results = []
promethee1_rank_data = []
promethee2_rank_data = []
electre3_rank_nfs_data = []
electre3_rank_nfs_pos_neg_data = []
electre3_median_rank_data = []
electre3_final_ranking_data = []
files_list = []

# Funkcja do wczytania określonych danych z plików JSON
def load_data_from_json(file_path, key):
    try:
        with open(file_path, 'r') as file:
            data = json.load(file)
            return data.get(key)
    except FileNotFoundError:
        print(f"Plik nie znaleziony: {file_path}")
        return None
c =0
# Przejście przez katalog z danymi wejściowymi
for root, dirs, files in os.walk(input_dir):
    for file in files:
        if c % 100 == 0:
            print(c, end="\r")
        if file.endswith('.json'):
            input_file_path = os.path.join(root, file)
            # Budowanie ścieżki do pliku wynikowego
            relative_path = os.path.relpath(root, input_dir)
            results_file_path = os.path.join(results_dir, relative_path, file)
            results_file_path = results_file_path.replace('data_', 'result_data_')
            files_list.append(results_file_path)  # Zapisanie nazwy pliku
            # Wczytanie danych wejściowych

            electre_binary_matrix = load_data_from_json(input_file_path, 'electre_binary_matrix')
            if electre_binary_matrix is not None:
                electre_binary_matrix_data.append(electre_binary_matrix)

            promethee_matrix = load_data_from_json(input_file_path, 'promethee_matrix')
            if promethee_matrix is not None:
                promethee_matrix_data.append(promethee_matrix)

            electre_matrix = load_data_from_json(input_file_path, 'electre_matrix')
            if electre_matrix is not None:
                electre_matrix_data.append(electre_matrix)

            promethee1_rank = load_data_from_json(input_file_path, 'promethee1_rank')
            if promethee1_rank is not None:
                promethee1_rank_data.append(promethee1_rank)

            promethee2_rank = load_data_from_json(input_file_path, 'promethee2_rank')
            if promethee2_rank is not None:
                promethee2_rank_data.append(promethee2_rank)

            electre3_rank_nfs = load_data_from_json(input_file_path, 'electre3_rank_nfs')
            if electre3_rank_nfs is not None:
                electre3_rank_nfs_data.append(electre3_rank_nfs)

            electre3_rank_nfs_pos_neg = load_data_from_json(input_file_path, 'electre3_rank_nfs_pos_neg')
            if electre3_rank_nfs_pos_neg is not None:
                electre3_rank_nfs_pos_neg_data.append(electre3_rank_nfs_pos_neg)

            electre3_median_rank = load_data_from_json(input_file_path, 'electre3_median_rank')
            if electre3_median_rank is not None:
                electre3_median_rank_data.append(electre3_median_rank)

            electre3_final_ranking = load_data_from_json(input_file_path, 'electre3_final_ranking')
            if electre3_final_ranking is not None:
                electre3_final_ranking_data.append(electre3_final_ranking)
            
            # Wczytanie wyników
            electre_binary_complete_data = load_data_from_json(results_file_path, 'electre_binary_complete_results')
            if electre_binary_complete_data is not None:
                electre_binary_complete_results.append(electre_binary_complete_data)

            electre_binary_partial_data = load_data_from_json(results_file_path, 'electre_binary_partial_results')
            if electre_binary_partial_data is not None:
                electre_binary_partial_results.append(electre_binary_partial_data)
            
            electre_complete_data = load_data_from_json(results_file_path, 'electre_complete_results')
            if electre_complete_data is not None:
                electre_complete_results.append(electre_complete_data)

            electre_partial_data = load_data_from_json(results_file_path, 'electre_partial_results')
            if electre_partial_data is not None:
                electre_partial_results.append(electre_partial_data)
            
            promethee_complete_data = load_data_from_json(results_file_path, 'promethee_complete_results')
            if promethee_complete_data is not None:
                promethee_complete_results.append(promethee_complete_data)

            promethee_partial_data = load_data_from_json(results_file_path, 'promethee_partial_results')
            if promethee_partial_data is not None:
                promethee_partial_results.append(promethee_partial_data)
            
        c += 1

print("Wczytano dane 'electre_binary_matrix_data':", len(electre_binary_matrix_data))
print("Wczytano dane 'promethee_matrix_data':", len(promethee_matrix_data))
print("Wczytano dane 'electre_matrix_data':", len(electre_matrix_data))
print("Wczytano dane 'promethee1_rank_data':", len(promethee1_rank_data))
print("Wczytano dane 'promethee2_rank_data':", len(promethee2_rank_data))
print("Wczytano dane 'electre3_rank_nfs_data':", len(electre3_rank_nfs_data))
print("Wczytano dane 'electre3_rank_nfs_pos_neg_data':", len(electre3_rank_nfs_pos_neg_data))
print("Wczytano dane 'electre3_median_rank_data':", len(electre3_median_rank_data))
print("Wczytano dane 'electre3_final_ranking_data':", len(electre3_final_ranking_data))

print("Wczytano dane 'electre_binary_complete_results':", len(electre_binary_complete_results))
print("Wczytano dane 'electre_binary_partial_results':", len(electre_binary_partial_results))
print("Wczytano dane 'electre_complete_results':", len(electre_complete_results))
print("Wczytano dane 'electre_partial_results':", len(electre_partial_results))
print("Wczytano dane 'promethee_complete_results':", len(promethee_complete_results))
print("Wczytano dane 'promethee_partial_results':", len(promethee_partial_results))
```

```python
def parse_filename(filename):
    parts = filename.split('\\')
    # Extracting parts of the filename
    criteria = parts[1].split('_')[1]
    alternatives = parts[2].split('_')[1]
    level = parts[3].split('_')[1]
    cutoff = parts[4].split('_')[1]
    dataset_number = parts[5].split('_')[2] #.split('_')[0]  # Split and take the first part

    return {
        'criteria': criteria,
        'alternatives': alternatives,
        'level': level,
        'cutoff': cutoff,
        'dataset_number': dataset_number
    }

# Applying the parsing function to each filename
data = [parse_filename(filename) for filename in files_list]

# Creating a DataFrame
df = pd.DataFrame(data)
```
